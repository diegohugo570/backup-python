# **Operações matemáticas básicas com Python**

A ordem das operações segue o mesmo conceito matematico. Se quer priorizar algum calculo, precisa colocar entre ().

- Soma
- Subtração
- Multiplicação
- Divisão
- Potência
- Mod % (resto da divisão)
---

```python
# Soma
print(10 + 12)
print(11 + 33)
```

```python
# Subtração
print(12-6)
```

```python
# Divisão com casas decimais: uma /
print(10 / 3 )
```

```python
# Divisão sem casas decimais: duas //
print(10 // 3)
```

```python
# Potência
print(2 ** 3)
```

```python
# Mod: resto da divisão
print(10 % 3)
```

```python
10 + (2-3)
```
