```python
import requests
from bs4 import BeautifulSoup

# URL da pÃ¡gina inicial
url = "https://books.toscrape.com/"

# Fazer a requisiÃ§Ã£o
response = requests.get(url)
response.encoding = 'utf-8'

# Verificar se a resposta foi bem-sucedida
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    livros = soup.find_all('article', class_='product_pod')

    for livro in livros:
        # Nome do livro
        nome = livro.h3.a['title']

        # PreÃ§o do livro
        preco = livro.find('p', class_='price_color').text.strip()

        # Nota (nÃºmero de estrelas)
        estrelas_tag = livro.find('p', class_='star-rating')
        classes = estrelas_tag.get('class')
        estrelas = classes[1]  # Ex: 'One', 'Two', 'Three'...

        # Converter texto para nÃºmero de estrelas
        conversao_estrelas = {
            'One': 1,
            'Two': 2,
            'Three': 3,
            'Four': 4,
            'Five': 5
        }
        nota = conversao_estrelas.get(estrelas, 0)

        print(f"ğŸ“˜ Livro: {nome}")
        print(f"ğŸ’° PreÃ§o: {preco}")
        print(f"â­ Nota: {nota} estrelas\n")

else:
    print("Erro ao acessar a pÃ¡gina:", response.status_code)
```

```python
# requests.text -> ver o que ele puxou
print(resposta)
```

```python
site = BeautifulSoup(resposta.text, "html.parser")
```

```python
dados = site.find_all("div", attrs= {"class" "i-search-result__content-wrapper"})
```

```python
produto = []
valor = []

for x in dados:
  nome = x.find_all("h2")[0].text
  produto.append(nome)
  print(nome)

  preco = x.find_all("span", attrs = {"class": "andes-money-amount__fraction"})[0].text
  valor.append(preco)
  print(preco)
```
