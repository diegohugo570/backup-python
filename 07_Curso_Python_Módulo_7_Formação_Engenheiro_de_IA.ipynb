### **2.2 Boxplot (Ca√ßando Outliers)**

O **Boxplot (Diagrama de Caixa)** √© uma ferramenta estat√≠stica usada para visualizar a distribui√ß√£o dos dados e identificar **outliers** de forma objetiva.

Ele √© constru√≠do a partir dos **quartis**:

* **Primeiro quartil (Q1):**  
  Valor abaixo do qual est√£o 25% dos dados.

* **Segundo quartil (Q2 ‚Äì Mediana):**  
  Valor central da distribui√ß√£o (50% dos dados abaixo e 50% acima).

* **Terceiro quartil (Q3):**  
  Valor abaixo do qual est√£o 75% dos dados.

A **Amplitude Interquartil (IQR)** √© definida por:
  
$$
\text{IQR} = Q3 - Q1
$$

Os **bigodes** do boxplot s√£o calculados como:

$$
\text{Limite inferior} = Q1 - 1.5 \cdot \text{IQR}
$$

$$
\text{Limite superior} = Q3 + 1.5 \cdot \text{IQR}
$$

Qualquer ponto que satisfa√ßa:

$$
x < Q1 - 1.5 \cdot \text{IQR}
\quad \text{ou} \quad
x > Q3 + 1.5 \cdot \text{IQR}
$$

√© classificado como **outlier** e deve ser analisado individualmente.
