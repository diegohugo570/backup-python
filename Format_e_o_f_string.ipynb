#1. Usando format() para personalizar o print:<br>
O método format() permite inserir variáveis ou valores diretamente dentro de uma string, facilitando a criação de mensagens dinâmicas.

```python
# Funciona em versões mais antigas do Python, antes do 3.6.
texto = "iago.braz@dascia.com"
nome = "Iago"
print("O e-mail de {} é {}".format(nome, texto))
```

#2. Usando f-string (Python 3.6+) para personalizar o print:<br>
As f-strings são uma maneira mais moderna e eficiente de formatar strings em Python. Elas permitem a inserção de variáveis diretamente na string, utilizando {} e colocando o prefixo f antes da string.

```python
# Aqui, dentro da f-string, as variáveis são inseridas diretamente entre chaves {} e a string é automaticamente formatada.
# Mais simples, mais rápida e legível (recomendada para Python 3.6+)
texto = "iago.braz@dascia.com"
nome = "Iago"
print(f"O e-mail de {nome} é {texto}")
```

# Usando %.nf para formatar o número com n casas decimais:
O %.2f é um exemplo de formatação que limita o número de casas decimais a 2. O número após o ponto (2) é o número de casas decimais que queremos.

```python
#format
valor = 3.14159265359
print("{:.4f}".format(valor))
```

```python
#f-string
valor = 3.14159265359
print(f"{valor : .3f}")
```
