1. Convertendo um inteiro para float:

```python
# Inicialmente, temos um inteiro
numero_inteiro = 10
type(numero_inteiro)

# Convertendo para float
numero_float = float(numero_inteiro)
print(type(numero_float))
print(numero_float)
```

2. Convertendo um float para inteiro:

```python
# Inicialmente, temos um float
numero_float = 3.98
type(numero_float)

# Convertendo para inteiro (truncando a parte decimal)
numero_inteiro = int(numero_float)
print(type(numero_inteiro))
print(numero_inteiro)
```

3. Convertendo um número para string:

```python
# Inicialmente, temos um inteiro
numero_inteiro = 100
type(numero_inteiro)
# Convertendo para string
numero_string = str(numero_inteiro)
print(type(numero_string))
print(numero_string)
```

4. Convertendo uma string para inteiro:

```python
# Inicialmente, temos uma string
numero_string = "250"

# Convertendo para inteiro
numero_inteiro = int(numero_string)

type(numero_inteiro)
print(numero_inteiro)
```

5. Convertendo uma string para float:

```python
# Inicialmente, temos uma string representando um número com ponto flutuante
numero_string = "3.14159"

# Convertendo para float
numero_float = float(numero_string)

print(numero_float)
print(type(numero_float))
```
