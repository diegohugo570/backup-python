# **Variaveis**
 Estrutura:
 nome_da_variavel = comando/valor
- É um nome que daremos para algum valor especifico
- Armazena uma informação importante que sera usada outras vezes durante nosso código
- Não pode ter caractere especial e espaço no nome. Se tiver espaço, coloque **_**: faturamento_total

**OBS**: Sempre execute a variavel para conseguir utiliza-la posteriormente

- tipos de variaveis:

  *   string: texto
  *   int: inteiro
  *   float: números com casas decimais
  *   bool ou boolean: True ou False
  *   Cuidado com os nomes restritos(Arquivo Download)
  *   **Variaveis em python não sao declaradas explicitamente, mas isso nao significa que voce deve ficar mudando o tempo todo**

- **Variaveis são Objetos:** Possuem caracteristicas e metodos especificos dependendo do tipo de variavel que você escolher.
